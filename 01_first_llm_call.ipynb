```python
import os
from dotenv import load_dotenv
load_dotenv()
print("Keys loaded successfully")
```

## Setting up the LLM
We use ChatGroq to access the Llama 3.3 70B model via Groq's free tier API.

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7,
    api_key=os.getenv("GROQ_API_KEY")
)
print("LLM initialized")
```

## Building the Chain
A LangChain chain connects a prompt template → LLM → output parser using the `|` pipe operator.

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI engineering assistant."),
    ("user", "{question}")
])

chain = prompt | llm | StrOutputParser()
print("Chain built")
```

```python
response = chain.invoke({"question": "What is an AI agent and why does it matter?"})
print(response)
```
